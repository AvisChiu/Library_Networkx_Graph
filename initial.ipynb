```python
import networkx as nx
import matplotlib.pyplot as plt

g = nx.cycle_graph(10)
x = nx.adj_matrix(g)
# print(x.todense())
nx.draw_networkx(g)
```

```python
n = 10
adj_mat = [(i, (i+1)%n) for i in range(n)]
adj_mat += [(i, (i+2)%n) for i in range(n)]
g = nx.Graph(adj_mat)
nx.draw_networkx(g)
```
